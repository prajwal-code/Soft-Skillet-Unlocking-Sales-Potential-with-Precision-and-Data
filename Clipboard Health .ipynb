```python
import pandas as pd
import os
import shutil

# Load Excel Data
file_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Soft Skillet_ March Data Live (1).xlsx"
df = pd.read_excel(file_path)

# Data Cleaning
connected_calls = df[df["Call Connect"] == "Connected"].copy()
connected_calls["Disposition"] = connected_calls["Disposition"].str.strip().str.title()

# Engagement Analysis by Pitch and Role
pivot_pitch_disposition = pd.crosstab(connected_calls["Pitch"], connected_calls["Disposition"])
pivot_role_pitch = pd.crosstab([connected_calls["Person"], connected_calls["Pitch"]], connected_calls["Disposition"])

# Save results to Excel
output_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Analysis_Results.xlsx"
with pd.ExcelWriter(output_path) as writer:
    pivot_pitch_disposition.to_excel(writer, sheet_name='Pitch_Disposition')
    pivot_role_pitch.to_excel(writer, sheet_name='Role_Pitch_Disposition')

print("Analysis results saved to Analysis_Results.xlsx")

# Organize Call Recordings
recordings_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Soft Skillet Call Recordings"
organized_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\organized_recordings"
os.makedirs(organized_path, exist_ok=True)

# Map Activity IDs to filenames
activity_ids = connected_calls["Activity ID"].astype(str).tolist()

for filename in os.listdir(recordings_path):
    for activity_id in activity_ids:
        if activity_id in filename:
            disposition = connected_calls.loc[connected_calls["Activity ID"] == int(activity_id), "Disposition"].values[0]
            pitch = connected_calls.loc[connected_calls["Activity ID"] == int(activity_id), "Pitch"].values[0]

            dest_folder = os.path.join(organized_path, f"{disposition}_{pitch}")
            os.makedirs(dest_folder, exist_ok=True)

            shutil.copy(os.path.join(recordings_path, filename), dest_folder)
            print(f"Copied {filename} to {dest_folder}")

print("Call recordings have been organized.")
```

```python
import pandas as pd
import os
import shutil

# Load Excel Data
file_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Soft Skillet_ March Data Live (1).xlsx"
df = pd.read_excel(file_path)

# Data Cleaning
connected_calls = df[df["Call Connect"] == "Connected"].copy()
connected_calls["Disposition"] = connected_calls["Disposition"].str.strip().str.title()

# Engagement Analysis by Pitch and Role
pivot_pitch_disposition = pd.crosstab(connected_calls["Pitch"], connected_calls["Disposition"])
pivot_role_pitch = pd.crosstab([connected_calls["Person"], connected_calls["Pitch"]], connected_calls["Disposition"])

# Save results to Excel
output_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Analysis_Results.xlsx"
with pd.ExcelWriter(output_path) as writer:
    pivot_pitch_disposition.to_excel(writer, sheet_name='Pitch_Disposition')
    pivot_role_pitch.to_excel(writer, sheet_name='Role_Pitch_Disposition')

print("Analysis results saved to Analysis_Results.xlsx")

# Organize Call Recordings
recordings_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\Soft Skillet Call Recordings"
organized_path = r"C:\Users\Nimith Narapareddy\Documents\Python Scripts\organized_recordings"
os.makedirs(organized_path, exist_ok=True)

# Ensure Activity IDs are integers and map them safely
activity_id_map = {str(activity_id): (disposition, pitch) for activity_id, disposition, pitch in zip(connected_calls["Activity ID"], connected_calls["Disposition"], connected_calls["Pitch"])}

for filename in os.listdir(recordings_path):
    matched_activity_id = next((aid for aid in activity_id_map if aid in filename), None)
    if matched_activity_id:
        disposition, pitch = activity_id_map[matched_activity_id]

        dest_folder = os.path.join(organized_path, f"{disposition}_{pitch}")
        os.makedirs(dest_folder, exist_ok=True)

        shutil.copy(os.path.join(recordings_path, filename), dest_folder)
        print(f"Copied {filename} to {dest_folder}")

print("Call recordings have been organized.")
```
